```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as inte
```

```python
t1 = np.linspace(0, 20, 101)
def lotka_plots(r, a, m, b):
    def lotka_volterra(r1, t):
        x, y = r1
        drdt = [r*x-a*x*y, b*x*y-m*y]
        return drdt
    r0 = [1.0, 1.0]
    sol = inte.odeint(lotka_volterra, r0, t1 )
    #plt.plot(t1, sol[:, 0], 'b', label='Prey population')
    #plt.plot(t1, sol[:, 1], 'g', label='Predator population')
    plt.plot(sol[:, 0], sol[:, 1])
    #plt.legend(loc='best')
    plt.xlabel('t')
    plt.grid()
    plt.show()
    
lotka_plots(1, 0.3, 1, 0.2)
```
